```python
import sqlite3

# Open the connection (this creates the database if there is not one already)
conn = sqlite3.connect('omada.db')
# Create cursor
c = conn.cursor()

# Drop all the tables if they exist
c.execute('''drop table if exists participant_accounts''')
c.execute('''drop table if exists weights''')
c.execute('''drop table if exists groups''')
c.execute('''drop table if exists group_message_board_posts''')
c.execute('''drop table if exists coach_accounts''')

# Create each table
c.execute(
    '''
    CREATE TABLE IF NOT EXISTS participant_accounts(
    id integer,
    email text,
    first_name text,
    last_name text,
    gender text,
    company text,
    group_id text,
    state text,
    city text,
    initial_weight_value double,
    height_in_inches double)
    '''    
)
c.execute(
    '''
    CREATE TABLE IF NOT EXISTS weights(
    id integer,
    participant_account_id integer,
    weighed_at timestamp,
    value double,
    filtered boolean
    )
    '''    
)
c.execute(
    '''
    CREATE TABLE IF NOT EXISTS groups(
    id integer,
    coach_account_id integer,
    start_date date
    )
    '''    
)
c.execute(
    '''
    CREATE TABLE IF NOT EXISTS group_message_board_posts(
    id integer,
    group_id integer,
    message text,
    created_at timestamp,
    created_by_account_type text,
    created_by_account_id integer
    )
    '''    
)
c.execute(
    '''
    CREATE TABLE IF NOT EXISTS coach_accounts(
    id integer,
    email text,
    first_name text,
    last_name text,
    state text,
    coach_manager_id integer
    )
    '''    
)
# Load the data for each table from csv files
# The CSV file is opened and then turned into a python list, which is then loaded by line
with open("C:\\Users\\cjbro\\Desktop\\Projects\\Omada Interview\\participants.csv", 'rb') as participants:
    reader = csv.reader(participants)
    part = list(reader)
    for line in part:
        c.execute('''insert into participant_accounts values(?,?,?,?,?,?,?,?,?,?,?)''', line)

with open("C:\\Users\\cjbro\\Desktop\\Projects\\Omada Interview\\weights.csv", 'rb') as weights:
    reader = csv.reader(weights)
    part = list(reader)
    for line in part:
        c.execute('''insert into weights values(?,?,?,?,?)''', line)

with open("C:\\Users\\cjbro\\Desktop\\Projects\\Omada Interview\\groups.csv", 'rb') as groups:
    reader = csv.reader(groups)
    part = list(reader)
    for line in part:
        c.execute('''insert into groups values(?,?,?)''', line)

with open("C:\\Users\\cjbro\\Desktop\\Projects\\Omada Interview\\group_message.csv", 'rb') as messages:
    reader = csv.reader(messages)
    part = list(reader)
    for line in part:
        c.execute('''insert into group_message_board_posts values(?,?,?,?,?,?)''', line)
        
with open("C:\\Users\\cjbro\\Desktop\\Projects\\Omada Interview\\coach.csv", 'rb') as coaches:
    reader = csv.reader(coaches)
    part = list(reader)
    for line in part:
        c.execute('''insert into coach_accounts values(?,?,?,?,?,?)''', line)

# Function for grabbing all the data from a table and printing it        
def table_results_return(table_name):
    c.execute("select * from "+table_name)
    results = c.fetchall()
    column_width=6
    for row in results:             
        for el in row:         
            print str(el).ljust(column_width), 
        print

# List of tables to run through table results function
table_list = ("participant_accounts","weights","groups","group_message_board_posts","coach_accounts")

# Go through table list and print all data for each table to make sure it all loaded
for item in table_list:
    table_results_return(item)

# Commit database changes and close the connection        
conn.commit()
conn.close()
```
